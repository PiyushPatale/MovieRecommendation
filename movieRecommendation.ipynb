```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv('movies.csv')
credits_df = pd.read_csv('credits.csv')
```

```python
movies_df.head()
```

```python
credits_df.head()
```

```python
movies = movies_df.merge(credits_df,on="title")
movies
```

```python
movies.info()
```

```python
movies = movies[['id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew', 'production_companies']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(o):
    list = []
    for i in ast.literal_eval(o):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(o):
    list = []
    j=0
    for i in ast.literal_eval(o):
        if(j!=5):
            list.append(i['name'])
            j+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
def convert3(o):
    list = []
    for i in ast.literal_eval(o):
        if(i['job'] == 'Director'):
            list.append(i['name'])
    return list
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
def convert4(o):
    list = []
    j=0
    for i in ast.literal_eval(o):
        if(j!=1):
            list.append(i['name'])
            j+=1
        else:
            break
    return list
```

```python
movies['production_companies'] = movies['production_companies'].apply(convert4)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew'] + movies['production_companies']
```

```python
movies = movies[['id', 'title', 'tags']]
```

```python
movies['tags'] = movies['tags'].apply(lambda x:[i.replace(" ", "") for i in x]) 
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
movies.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stemWord(text):
    t = []

    for i in text.split():
        t.append(ps.stem(i))

    return " ".join(t)
```

```python
movies
```

```python
movies['tags'] = movies['tags'].apply(stemWord)
```

```python
movies
```

```python
#Vectorisation
from sklearn.feature_extraction.text import CountVectorizer

countVec = CountVectorizer(max_features = 6000, stop_words = 'english')

arr = countVec.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim = cosine_similarity(arr)
```

```python
sim
```

```python
# countVec.get_feature_names()
```

```python
def recommend(movie):
    movie_idx = movies[movies['title'] == movie].index[0]
    distance = sim[movie_idx]
    finalMovies = sorted(list(enumerate(distance)),reverse = True, key = lambda x: x[1])[1:6]

    for i in finalMovies:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
